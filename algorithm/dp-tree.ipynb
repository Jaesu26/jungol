# 트리에서의 동적 계획법 (Tree DP)

`-` 트리에서 동적 계획법을 사용해보자

## 가지치기

- 문제 출처: [정올 1563번](https://jungol.co.kr/problem/1563)

`-` 어떤 노드 $u$의 자식들 $c_1, c_2, \cdots, c_n$을 고려하자. 그리고 $W_{u,v}$를 노드 $u$와 $v$를 잇는 간선의 가중치라고 하자. 그럼 다음을 만족한다

`-` $\operatorname{dp}[u] = \sum\limits_{i=1}^{n}\min\left(\operatorname{dp}[c_i], W_{u, c_i}\right)$, (단, $u$가 리프 노드인 경우 $\operatorname{dp}[u]=\infty$)

`-` $u$와 연결된 노드가 하나뿐이고 루트 노드가 아니라면 $u$는 리프 노드이다

`-` 로직상 각 노드를 $1$번만 방문하므로 전체 알고리즘의 시간 복잡도는 $O(N)$이다

```python
def dfs(tree, node, parent, root, dp):
    min_total_weight = 0
    if len(tree[node]) == 1 and node != root:
        dp[node] = INF
        return dp[node]
    for child, edge_weight in tree[node]:
        if child == parent:
            continue
        dp_child = dfs(tree, child, node, root, dp)
        min_total_weight += min(dp_child, edge_weight)
    dp[node] = min_total_weight
    return dp[node]


def solve_testcase(root, n):
    tree = [[] for _ in range(n + 1)]
    for _ in range(n - 1):
        u, v, w = map(int, input().split())
        tree[u].append((v, w))
        tree[v].append((u, w))
    dp = [INF] * (n + 1)
    answer = dfs(tree, root, NONE, root, dp)
    return answer


def solution():
    global INF, NONE
    INF = float("inf")
    NONE = -1
    while True:
        N, R = map(int, input().split())
        if N == 0 and R == 0:
            break
        answer = solve_testcase(R, N)
        print(answer)


solution()

# input
# 0 0
```
