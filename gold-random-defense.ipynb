## 목걸이(The necklace of beads)

- 문제 출처: [정올 1114번](https://jungol.co.kr/problem/1114)

### 정답 풀이 (구현)

`-` 이거 왼쪽에 문제 역링크가 표시돼서 태그 등이 스포된다. 이 문제는 Advanced의 기타 알고리즘에 등재된 문제이다

`-` 새로고침 눌렀더니 문제의 난이도가 표시된다 (예?)

`-` 오케이 분석 완료... 기타 알고리즘이고 난이도는 골드 $5$

`-` 모든 게 망해버렸지만 그래도 풀긴 해야지. 다음부턴 왼쪽은 가리고 새로고침은 절대 누르지 않는 상태로 문제를 풀자

`-` 이 문제는 간단하게 해결할 수 있을 것 같다 (ㅋㅋ)

`-` 목걸이의 한 부분을 끊었다고 해보자

`-` $i$번째 유리알과 $i+1$번째 유리알 사이를 끊었다고 한다면 $i+1\sim n$번째 유리알과 $1\sim i$번째 유리알을 이어 붙여 새로운 목걸이로 만드는 것이다

`-` 첫 줄과 끝 줄에서 시작해 색이 계속되는 유리알을 모으는 건 $O(n)$이며 끊을 수 있는 경우의 수는 $O(n)$이다

`-` 따라서 $O\left(n^2\right)$에 문제를 해결할 수 있고 $n$은 최대 $350$이므로 제한 시간 안에 동작할 수 있다

`-` 참고로 색이 계속되는 유리알을 모을 때 흰색 알의 존재 때문에 같은 유리알을 중복해서 모을 수 있다

`-` 예를 들어 $1 \sim 200$번째 유리알까지 모으고 $100\sim 300$번째 유리알까지 모았다면 $100\sim 200$번째 유리알은 흰색 알이며 중복해서 셈해졌다

`-` 이러면 $n$개의 유리알을 모두 모을 수 있는 것이니 정답은 $n$이 된다

`-` $53$분 걸렸고 $5$번 시도만에 맞혔다 (모든 경우의 수를 제대로 고려하지 않았다, 너무 대충함) 

```python
def cut(array, i):
    return array[i:] + array[:i]


def collect(necklace):
    n = len(necklace)
    i_f = _collect_forward(necklace)
    i_b = _collect_backward(necklace)
    if i_f < i_b:
        return (i_f + 1) + (n - i_b)
    return n


def _collect_forward(necklace):
    v = necklace
    n = len(v)
    i = 0
    c = v[i]
    while i < n - 1 and (v[i] == v[i + 1] or v[i + 1] == W or c == W or (v[i] == W and v[i + 1] == c)):
        if c == W and v[i + 1] != W:
            c = v[i + 1]
        i += 1
    return i


def _collect_backward(necklace):
    v = necklace
    n = len(v)
    i = n - 1
    c = v[i]
    while i > 0 and (v[i] == v[i - 1] or v[i - 1] == W or c == W or (v[i] == W and v[i - 1] == c)):
        if c == W and v[i - 1] != W:
            c = v[i - 1]
        i -= 1
    return i


def solution():
    global W
    W = "w"
    _ = input()
    original_necklace = input().rstrip()
    n = len(original_necklace)
    answer = 0
    for i in range(n + 1):
        necklace = cut(original_necklace, i)
        count = collect(necklace)
        answer = max(count, answer)
    print(answer)


solution()

# input
# 5
# wbrbr
```

## 벌집 네트워크

- 문제 출처: [정올 8681번](https://jungol.co.kr/problem/8681)

### 정답 풀이 (그래프 탐색)

`-` $T$와 $N$의 크기가 작아 나이브하게 구현해도 된다 (파싱만 제대로 하자)

`-` 그래프의 노드와 간선 개수는 $O(N)$이므로 DFS의 시간 복잡도는 $O(N)$이다 (BFS도 상관없다, 뭐든 탐색만 제대로 하면 됨)

`-` 이를 $N$번 반복하면 총점을 알 수 있고 총 $T$번 반복해야 하므로 전체 알고리즘의 시간 복잡도는 $O\left(TN^2\right)$이다

`-` $34$분 걸렸고 $2$트에 맞혔다. rstrip 함수를 빼먹지 말자

```python
def compute_total_score(graph, type_, alphabet, alphabet2index):
    global SIZE
    score = 0
    for u in graph:
        SIZE = 0
        dfs(graph, u, {}, type_, alphabet, alphabet2index)
        score += SIZE
    return score


def dfs(graph, u, visited, type_, alphabet, alphabet2index):
    global SIZE
    SIZE += 1
    visited[u] = True
    next_nodes = get_next_nodes(u, alphabet, alphabet2index)
    for v in next_nodes:
        if v in graph and v not in visited and graph[v][type_] == graph[u][type_]:
            dfs(graph, v, visited, type_, alphabet, alphabet2index)


def get_next_nodes(node, alphabet, alphabet2index):
    col = node[0]
    row = node[1]
    row_int = int(row)
    level = node[2]
    next_nodes = []
    if level == "D":
        next_nodes.append(col + row + "U")
        i = alphabet2index[col]
        if i > 0:
            next_nodes.append(alphabet[i - 1] + row + "U")
        if row_int > 1:
            next_nodes.append(col + str(row_int - 1) + "U")
        return next_nodes
    next_nodes.append(col + row + "D")
    i = alphabet2index[col]
    if i < COL_SIZE - 1:
        next_nodes.append(alphabet[i + 1] + row + "D")
    if row_int < ROW_SIZE:
        next_nodes.append(col + str(row_int + 1) + "D")
    return next_nodes


def solve_testcase():
    global ROW_SIZE, COL_SIZE, COLOR, NATURE
    ROW_SIZE, COL_SIZE = 6, 6
    COLOR, NATURE = 0, 1
    N = int(input())
    alphabet = "abcdef"
    alphabet2index = {a: i for i, a in enumerate(alphabet)}
    graph = {}
    for _ in range(N):
        node, bee = input().split()
        graph[node] = bee
    jungol_score = compute_total_score(graph, NATURE, alphabet, alphabet2index)
    hangul_score = compute_total_score(graph, COLOR, alphabet, alphabet2index)
    print(jungol_score, hangul_score)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# 1
# 2
# c4U C2
# e4U C2
```

## 연속 수열 (JEDAN)

- 문제 출처: [정올 1690번](https://jungol.co.kr/problem/1690)

### 정답 풀이 (동적 계획법)

`-` $\operatorname{dp}[i][c]$를 수열의 $i$번째 항까지 고려했고 $C_i = c$일 때 연속 수열을 만들 수 있는 방법의 수라고 하자

`-` $C_i=-1$이라면 이는 빈칸이므로 가능한 모든 $c$에 대해 $\operatorname{dp}[i][c] = \operatorname{dp}[i - 1][c - 1] + \operatorname{dp}[i - 1][c] + \operatorname{dp}[i - 1][c + 1]$이다

`-` $C_i > 0$이라면 값이 정해져 있으므로 $\operatorname{dp}[i][C_i] = \operatorname{dp}[i - 1][C_i - 1] + \operatorname{dp}[i - 1][C_i] + \operatorname{dp}[i - 1][C_i + 1]$이다

`-` 단, 연속 수열의 모든 항은 음이 아닌 정수여야 하고 첫 항과 마지막 항은 $0$이어야 함에 유의하자

`-` 전체 알고리즘의 시간 복잡도는 전체가 빈칸인 경우 최악의 경우 $O\left(N^2\right)$이다

`-` 시간 초과가 발생했다

`-` $i$번째 원소를 고려할 때 $C_i \ge i$ 또는 $C_i > N - i$라면 연속 수열을 만들 수 없으니 실행을 중단하고 $0$을 출력하자

`-` 이번엔 메모리 초과가 발생했다

`-` $\operatorname{dp}[i]$는 $\operatorname{dp}[i - 1]$에만 의존하니 이를 이용해 메모리 사용량을 $O(N)$으로 줄이자. 메모리 사용량이 줄었으니 딕셔너리 대신 배열을 이용해서 표현하자

`-` $59$분 걸렸고 $4$번 만에 맞혔다

```python
def count_continuous_sequence(array, mod):
    if array[0] != BLANK and array[0] != 0:
        return 0
    n = len(array)
    dp = [0] * n
    dp[0] = 1
    for i, a in enumerate(array[1:], start=1):
        temp = [0] * n
        if a != BLANK:
            if a > i or a > n - i:
                return 0
            temp[a] += dp[a] + dp[a + 1] + (dp[a - 1] if a > 0 else 0)
            temp[a] %= mod
            dp = temp
            continue
        for a in range(min(i, n - i) + 1):
            temp[a] += dp[a] + dp[a + 1] + (dp[a - 1] if a > 0 else 0)
            temp[a] %= mod
        dp = temp
    return dp[0]


def solution():
    global BLANK
    BLANK = -1
    N = int(input())
    C = list(map(int, input().split()))
    mod = 10**9 + 7
    answer = count_continuous_sequence(C, mod)
    print(answer)


solution()

# input
# 1
# 0
```

## 줄서기

- 문제 출처: [정올 3075번](https://jungol.co.kr/problem/3075)

### 정답 풀이 (애드 혹)

`-` 처음에 문제를 완전히 잘못 이해해서 $40$분을 허비했다

`-` 그 후 갈피를 못잡고 세그먼트 트리를 쓰겠다고 해서 $30$분을 내다 버렸다

`-` 근데 생각해보니 다 필요 없었다 

`-` 각 학생별로 $X$로서 등장한 횟수와 $Y$로서 등장한 횟수를 기록하자

`-` 만약 $X_i$가 $0$이고 $Y_i$도 $0$이라면 학생$i$ 앞에는 $i-1$명이 존재하고 있으므로 $i$번 카드를 소지해야만 한다 (그렇지 않으면 $X_i=0,Y_i=0$이라는 조건에 모순됨)

`-` 하지만 그렇지 않다면 학생$i$가 가진 카드의 번호는 기존 $i$에서 $X_i$만큼 커져야 하며 $Y_i$만큼 작아져야 한다

`-` 따라서 학생$1$부터 학생$N$까지 카드를 $O(N)$에 배정해 줄 수 있다. 만약 이미 배정한 카드를 또 필요로 한다면 모순이니 $-1$을 출력하면 된다

`-` 모든 $i$에 대해 $X_i,Y_i$를 기록하는 건 $O(M)$의 시간 복잡도를 가지므로 전체 알고리즘의 시간 복잡도는 $O(N + M)$이다

`-` $1$시간 $42$분 걸렸고 $1$번에 맞혔다

```python
def find_order(x_counts, y_counts):
    n = len(x_counts) - 1
    order = [None] * n
    seen = set()
    for i in range(1, n + 1):
        j = i + x_counts[i] - y_counts[i]
        if j in seen:
            return [-1]
        seen.add(j)
        order[i - 1] = j
    return order


def solution():
    N, M = map(int, input().split())
    x_counts = [0] * (N + 1)
    y_counts = [0] * (N + 1)
    for _ in range(M):
        X, Y = map(int, input().split())
        x_counts[X] += 1
        y_counts[Y] += 1
    order = find_order(x_counts, y_counts)
    print(*order)


solution()

# input
# 1 0
```
