# 분할 정복 (Devide And Conquer)

`-` 하나의 문제를 여러 개의 하위 문제로 분할하여 해결한 뒤 병합하는 알고리즘

## 색종이 만들기3(압축해제)

- 문제 출처: [정올 3249번](https://jungol.co.kr/problem/3249)

`-` [색종이 만들기](https://www.acmicpc.net/problem/2630) 문제에선 색종이를 압축했다. 여기선 압축된 정보를 바탕으로 원래의 색종이로 복구해야 한다

`-` 압축 문자열의 문자 하나하나는 전체 색종이에서 일부 정사각형 영역의 정보를 나타낸다. 예로 시작 문자가 가리키는 영역의 정보는 전체 색종이이다

`-` 현재 문자가 가리키는 정사각형 영역의 좌상, 우하 좌표와 현재 문자가 $0$인지 $1$인지 $X$인지 알면 색종이를 복구할 수 있다

`-` $X$가 아니라면 함수를 종료하고 그렇지 않다면 현재 영역을 $4$등분 한 뒤 재귀적으로 함수를 호출하면 된다

`-` 알고리즘의 시간 복잡도는 $O\left(N^2\right)$이다

```python
def upzip(row_lu, col_lu, row_rd, col_rd, code, paper):
    global ITH
    ITH += 1
    if code[ITH] != "X":
        fill(paper, code[ITH], row_lu, col_lu, row_rd, col_rd)
        return
    row_mid = (row_rd + row_lu) // 2
    col_mid = (col_rd + col_lu) // 2
    upzip(row_lu, col_lu, row_mid, col_mid, code, paper)
    upzip(row_lu, col_mid + 1, row_mid, col_rd, code, paper)
    upzip(row_mid + 1, col_lu, row_rd, col_mid, code, paper)
    upzip(row_mid + 1, col_mid + 1, row_rd, col_rd, code, paper)


def fill(matrix, value, row_lu, col_lu, row_rd, col_rd):
    for row in range(row_lu, row_rd + 1):
        for col in range(col_lu, col_rd + 1):
            matrix[row][col] = value


def solution():
    global ITH
    ITH = -1
    N = int(input())
    code = input().rstrip()
    paper = [[None] * N for _ in range(N)]
    upzip(0, 0, N - 1, N - 1, code, paper)
    print(N)
    for row in paper:
        print(*row)


solution()

# 8
# 0
```

`-` 정올도 백준처럼 단계별로 문제가 나뉘어 있어서 되게 좋다

## 타일 채우기

- 문제 출처: [정올 2543번](https://jungol.co.kr/problem/2543)

`-` $\frac{3}{4}n^2$개의 조각으로 이루어진 회전시켜서 ㄴ 모양의 타일을 $T(n)$이라고 하자

`-` $T(n)$은 $4$개의 $T\left(\frac{n}{2}\right)$로 만들 수 있다

`-` 한 변의 길이가 $N$인 화장실 바닥을 고려하자. 이를 같은 크기의 정사각형 $4$개로 분할할 수 있다

`-` 이때 이 중 한 곳에는 배수구가 있을텐데 이를 제외한 나머지 $3$개의 정사각형을 합치면 $T(N)$이다

`-` $T(N)$을 기저 타일 $4$개로 채우는 것은 쉬우며 배수구가 존재하는 $1$개의 정사각형은 기존 문제에서 $N$이 $\frac{N}{2}$이 된 경우이다

`-` 따라서 이를 $N=2$일 때까지 반복하면 문제를 해결할 수 있다

`-` 참고로 큰 ㄴ 모양의 타일을 기저 타일로 채우는 함수와 화장실 바닥을 타일로 채우는 함수를 나누어서 관리해야 된다 

`-` 전체 알고리즘의 시간 복잡도는 $O\left(N^2\right)$이다

```python
def fill_floor(r_lu, c_lu, x, y, n, matrix):
    if n == 2:
        if r_lu == x and c_lu == y:
            fill_L1(r_lu + 1, c_lu + 1, n, matrix)
        if r_lu == x and c_lu + 1 == y:
            fill_L2(r_lu + 1, c_lu, n, matrix)
        if r_lu + 1 == x and c_lu == y:
            fill_L3(r_lu, c_lu + 1, n, matrix)
        if r_lu + 1 == x and c_lu + 1 == y:
            fill_L4(r_lu, c_lu, n, matrix)
        return
    m = n // 2
    if x < r_lu + m and y < c_lu + m:
        fill_floor(r_lu, c_lu, x, y, m, matrix)
        fill_L1(r_lu + n - 1, c_lu + n - 1, n, matrix)
    if x < r_lu + m and y >= c_lu + m:
        fill_floor(r_lu, c_lu + m, x, y, m, matrix)
        fill_L2(r_lu + n - 1, c_lu, n, matrix)
    if x >= r_lu + m and y < c_lu + m:
        fill_floor(r_lu + m, c_lu, x, y, m, matrix)
        fill_L3(r_lu, c_lu + n - 1, n, matrix)
    if x >= r_lu + m and y >= c_lu + m:
        fill_floor(r_lu + m, c_lu + m, x, y, m, matrix)
        fill_L4(r_lu, c_lu, n, matrix)


def fill_L1(r, c, n, matrix):
    if n == 2:
        matrix[r][c - 1] = 1
        matrix[r][c] = 1
        matrix[r - 1][c] = 1
        return
    m = n // 2
    k = m // 2
    fill_L1(r - m + k, c - m + k, m, matrix)
    fill_L1(r, c, m, matrix)
    fill_L2(r, c - n + 1, m, matrix)
    fill_L3(r - n + 1, c, m, matrix)


def fill_L2(r, c, n, matrix):
    if n == 2:
        matrix[r - 1][c] = 2
        matrix[r][c] = 2
        matrix[r][c + 1] = 2
        return
    m = n // 2
    k = m // 2
    fill_L2(r - m + k, c + m - k, m, matrix)
    fill_L2(r, c, m, matrix)
    fill_L4(r - n + 1, c, m, matrix)
    fill_L1(r, c + n - 1, m, matrix)


def fill_L3(r, c, n, matrix):
    if n == 2:
        matrix[r][c - 1] = 3
        matrix[r][c] = 3
        matrix[r + 1][c] = 3
        return
    m = n // 2
    k = m // 2
    fill_L3(r + m - k, c - m + k, m, matrix)
    fill_L3(r, c, m, matrix)
    fill_L4(r, c - n + 1, m, matrix)
    fill_L1(r + n - 1, c, m, matrix)


def fill_L4(r, c, n, matrix):
    if n == 2:
        matrix[r][c + 1] = 4
        matrix[r][c] = 4
        matrix[r + 1][c] = 4
        return
    m = n // 2
    k = m // 2
    fill_L4(r + m - k, c + m - k, m, matrix)
    fill_L4(r, c, m, matrix)
    fill_L2(r + n - 1, c, m, matrix)
    fill_L3(r, c + n - 1, m, matrix)


def solution():
    N = int(input())
    X, Y = map(int, input().split())
    matrix = [[0] * N for _ in range(N)]
    fill_floor(0, 0, X, Y, N, matrix)
    for row in matrix:
        print(*row)


solution()

# input
# 4
# 2 1
```
