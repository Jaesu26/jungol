## 목걸이(The necklace of beads)

- 문제 출처: [정올 1114번](https://jungol.co.kr/problem/1114)

### 정답 풀이 (구현)

`-` 이거 왼쪽에 문제 역링크가 표시돼서 태그 등이 스포된다. 이 문제는 Advanced의 기타 알고리즘에 등재된 문제이다

`-` 새로고침 눌렀더니 문제의 난이도가 표시된다 (예?)

`-` 오케이 분석 완료... 기타 알고리즘이고 난이도는 골드 $5$

`-` 모든 게 망해버렸지만 그래도 풀긴 해야지. 다음부턴 왼쪽은 가리고 새로고침은 절대 누르지 않는 상태로 문제를 풀자

`-` 이 문제는 간단하게 해결할 수 있을 것 같다 (ㅋㅋ)

`-` 목걸이의 한 부분을 끊었다고 해보자

`-` $i$번째 유리알과 $i+1$번째 유리알 사이를 끊었다고 한다면 $i+1\sim n$번째 유리알과 $1\sim i$번째 유리알을 이어 붙여 새로운 목걸이로 만드는 것이다

`-` 첫 줄과 끝 줄에서 시작해 색이 계속되는 유리알을 모으는 건 $O(n)$이며 끊을 수 있는 경우의 수는 $O(n)$이다

`-` 따라서 $O\left(n^2\right)$에 문제를 해결할 수 있고 $n$은 최대 $350$이므로 제한 시간 안에 동작할 수 있다

`-` 참고로 색이 계속되는 유리알을 모을 때 흰색 알의 존재 때문에 같은 유리알을 중복해서 모을 수 있다

`-` 예를 들어 $1 \sim 200$번째 유리알까지 모으고 $100\sim 300$번째 유리알까지 모았다면 $100\sim 200$번째 유리알은 흰색 알이며 중복해서 셈해졌다

`-` 이러면 $n$개의 유리알을 모두 모을 수 있는 것이니 정답은 $n$이 된다

`-` $53$분 걸렸고 $5$번 시도만에 맞혔다 (모든 경우의 수를 제대로 고려하지 않았다, 너무 대충함) 

````python
def cut(array, i):
    return array[i:] + array[:i]


def collect(necklace):
    n = len(necklace)
    i_f = _collect_forward(necklace)
    i_b = _collect_backward(necklace)
    if i_f < i_b:
        return (i_f + 1) + (n - i_b)
    return n


def _collect_forward(necklace):
    v = necklace
    n = len(v)
    i = 0
    c = v[i]
    while i < n - 1 and (v[i] == v[i + 1] or v[i + 1] == W or c == W or (v[i] == W and v[i + 1] == c)):
        if c == W and v[i + 1] != W:
            c = v[i + 1]
        i += 1
    return i


def _collect_backward(necklace):
    v = necklace
    n = len(v)
    i = n - 1
    c = v[i]
    while i > 0 and (v[i] == v[i - 1] or v[i - 1] == W or c == W or (v[i] == W and v[i - 1] == c)):
        if c == W and v[i - 1] != W:
            c = v[i - 1]
        i -= 1
    return i


def solution():
    global W
    W = "w"
    _ = input()
    original_necklace = input().rstrip()
    n = len(original_necklace)
    answer = 0
    for i in range(n + 1):
        necklace = cut(original_necklace, i)
        count = collect(necklace)
        answer = max(count, answer)
    print(answer)


solution()

# input
# 5
# wbrbr
````

## 벌집 네트워크

- 문제 출처: [정올 8681번](https://jungol.co.kr/problem/8681)

### 정답 풀이 (그래프 탐색)

`-` $T$와 $N$의 크기가 작아 나이브하게 구현해도 된다 (파싱만 제대로 하자)

`-` 그래프의 노드와 간선 개수는 $O(N)$이므로 DFS의 시간 복잡도는 $O(N)$이다 (BFS도 상관없다, 뭐든 탐색만 제대로 하면 됨)

`-` 이를 $N$번 반복하면 총점을 알 수 있고 총 $T$번 반복해야 하므로 전체 알고리즘의 시간 복잡도는 $O\left(TN^2\right)$이다

`-` $34$분 걸렸고 $2$트에 맞혔다. rstrip 함수를 빼먹지 말자

````python
def compute_total_score(graph, type_, alphabet, alphabet2index):
    global SIZE
    score = 0
    for u in graph:
        SIZE = 0
        dfs(graph, u, {}, type_, alphabet, alphabet2index)
        score += SIZE
    return score


def dfs(graph, u, visited, type_, alphabet, alphabet2index):
    global SIZE
    SIZE += 1
    visited[u] = True
    next_nodes = get_next_nodes(u, alphabet, alphabet2index)
    for v in next_nodes:
        if v in graph and v not in visited and graph[v][type_] == graph[u][type_]:
            dfs(graph, v, visited, type_, alphabet, alphabet2index)


def get_next_nodes(node, alphabet, alphabet2index):
    col = node[0]
    row = node[1]
    row_int = int(row)
    level = node[2]
    next_nodes = []
    if level == "D":
        next_nodes.append(col + row + "U")
        i = alphabet2index[col]
        if i > 0:
            next_nodes.append(alphabet[i - 1] + row + "U")
        if row_int > 1:
            next_nodes.append(col + str(row_int - 1) + "U")
        return next_nodes
    next_nodes.append(col + row + "D")
    i = alphabet2index[col]
    if i < COL_SIZE - 1:
        next_nodes.append(alphabet[i + 1] + row + "D")
    if row_int < ROW_SIZE:
        next_nodes.append(col + str(row_int + 1) + "D")
    return next_nodes


def solve_testcase():
    global ROW_SIZE, COL_SIZE, COLOR, NATURE
    ROW_SIZE, COL_SIZE = 6, 6
    COLOR, NATURE = 0, 1
    N = int(input())
    alphabet = "abcdef"
    alphabet2index = {a: i for i, a in enumerate(alphabet)}
    graph = {}
    for _ in range(N):
        node, bee = input().split()
        graph[node] = bee
    jungol_score = compute_total_score(graph, NATURE, alphabet, alphabet2index)
    hangul_score = compute_total_score(graph, COLOR, alphabet, alphabet2index)
    print(jungol_score, hangul_score)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# 1
# 2
# c4U C2
# e4U C2
````

## 연속 수열 (JEDAN)

- 문제 출처: [정올 1690번](https://jungol.co.kr/problem/1690)

### 정답 풀이 (동적 계획법)

`-` $\operatorname{dp}[i][c]$를 수열의 $i$번째 항까지 고려했고 $C_i = c$일 때 연속 수열을 만들 수 있는 방법의 수라고 하자

`-` $C_i=-1$이라면 이는 빈칸이므로 가능한 모든 $c$에 대해 $\operatorname{dp}[i][c] = \operatorname{dp}[i - 1][c - 1] + \operatorname{dp}[i - 1][c] + \operatorname{dp}[i - 1][c + 1]$이다

`-` $C_i > 0$이라면 값이 정해져 있으므로 $\operatorname{dp}[i][C_i] = \operatorname{dp}[i - 1][C_i - 1] + \operatorname{dp}[i - 1][C_i] + \operatorname{dp}[i - 1][C_i + 1]$이다

`-` 단, 연속 수열의 모든 항은 음이 아닌 정수여야 하고 첫 항과 마지막 항은 $0$이어야 함에 유의하자

`-` 전체 알고리즘의 시간 복잡도는 전체가 빈칸인 경우 최악의 경우 $O\left(N^2\right)$이다

`-` 시간 초과가 발생했다

`-` $i$번째 원소를 고려할 때 $C_i \ge i$ 또는 $C_i > N - i$라면 연속 수열을 만들 수 없으니 실행을 중단하고 $0$을 출력하자

`-` 이번엔 메모리 초과가 발생했다

`-` $\operatorname{dp}[i]$는 $\operatorname{dp}[i - 1]$에만 의존하니 이를 이용해 메모리 사용량을 $O(N)$으로 줄이자. 메모리 사용량이 줄었으니 딕셔너리 대신 배열을 이용해서 표현하자

`-` $59$분 걸렸고 $4$번 만에 맞혔다

````python
def count_continuous_sequence(array, mod):
    if array[0] != BLANK and array[0] != 0:
        return 0
    n = len(array)
    dp = [0] * n
    dp[0] = 1
    for i, a in enumerate(array[1:], start=1):
        temp = [0] * n
        if a != BLANK:
            if a > i or a > n - i:
                return 0
            temp[a] += dp[a] + dp[a + 1] + (dp[a - 1] if a > 0 else 0)
            temp[a] %= mod
            dp = temp
            continue
        for a in range(min(i, n - i) + 1):
            temp[a] += dp[a] + dp[a + 1] + (dp[a - 1] if a > 0 else 0)
            temp[a] %= mod
        dp = temp
    return dp[0]


def solution():
    global BLANK
    BLANK = -1
    N = int(input())
    C = list(map(int, input().split()))
    mod = 10**9 + 7
    answer = count_continuous_sequence(C, mod)
    print(answer)


solution()

# input
# 1
# 0
````

## 줄서기

- 문제 출처: [정올 3075번](https://jungol.co.kr/problem/3075)

### 정답 풀이 (애드 혹)

`-` 처음에 문제를 완전히 잘못 이해해서 $40$분을 허비했다

`-` 그 후 갈피를 못잡고 세그먼트 트리를 쓰겠다고 해서 $30$분을 내다 버렸다

`-` 근데 생각해보니 다 필요 없었다 

`-` 각 학생별로 $X$로서 등장한 횟수와 $Y$로서 등장한 횟수를 기록하자

`-` 만약 $X_i$가 $0$이고 $Y_i$도 $0$이라면 학생$i$ 앞에는 $i-1$명이 존재하고 있으므로 $i$번 카드를 소지해야만 한다 (그렇지 않으면 $X_i=0,Y_i=0$이라는 조건에 모순됨)

`-` 하지만 그렇지 않다면 학생$i$가 가진 카드의 번호는 기존 $i$에서 $X_i$만큼 커져야 하며 $Y_i$만큼 작아져야 한다

`-` 따라서 학생$1$부터 학생$N$까지 카드를 $O(N)$에 배정해 줄 수 있다. 만약 이미 배정한 카드를 또 필요로 한다면 모순이니 $-1$을 출력하면 된다

`-` 모든 $i$에 대해 $X_i,Y_i$를 기록하는 건 $O(M)$의 시간 복잡도를 가지므로 전체 알고리즘의 시간 복잡도는 $O(N + M)$이다

`-` $1$시간 $42$분 걸렸고 $1$번에 맞혔다

````python
def find_order(x_counts, y_counts):
    n = len(x_counts) - 1
    order = [None] * n
    seen = set()
    for i in range(1, n + 1):
        j = i + x_counts[i] - y_counts[i]
        if j in seen:
            return [-1]
        seen.add(j)
        order[i - 1] = j
    return order


def solution():
    N, M = map(int, input().split())
    x_counts = [0] * (N + 1)
    y_counts = [0] * (N + 1)
    for _ in range(M):
        X, Y = map(int, input().split())
        x_counts[X] += 1
        y_counts[Y] += 1
    order = find_order(x_counts, y_counts)
    print(*order)


solution()

# input
# 1 0
````

## 두수의 합2

- 문제 출처: [정올 1353번](https://jungol.co.kr/problem/1353)

### 정답 풀이 (해시를 사용한 집합과 맵)

`-` 같은 수가 중복되는 경우는 $1$개로 처리함에 유의하자

`-` 배열을 기반으로 카운터 딕셔너리를 만들고 이를 순회할 것이다. 비어있는 집합을 하나 만들어두고 여기에 합이 $M$이 되는 쌍을 넣자

`-` 딕셔너리의 키 값 $x$에 대해 $M-x$가 딕셔너리에 없으면 무시해야 한다. 또한, $x = M-x$인 경우 배열에서 $x$가 등장한 횟수가 $1$이라면 무시해야 한다

`-` 이 외의 경우엔 $M-x$가 딕셔너리에 존재하면 튜플 $(\min(x, M- x), \max(x, M-x))$를 집합에 넣자 (이미 존재하면 생략)

`-` 그럼 최종적으로 집합의 크기가 정답이 된다

`-` 전체 알고리즘의 시간 복잡도는 최악의 경우 $O(N)$이다

`-` $21$분 걸렸고 $2$번 만에 맞혔다

`-` 처음에 조건 분기를 이상하게 구현해서 틀렸었다. 먼저 $M-x$가 딕셔너리에 없으면 continue 문을 사용해 다음 $x$로 넘어갔다. 그 후 $x=M-x$이고 등장 횟수가 $2$ 이상이면 카운팅을 $1$ 올리고 continue 문을 사용해 다음 $x$로 넘어갔다. 이제 $M-x$가 딕셔너리에 존재하기만 하면 카운팅을 $1$ 올리고 다음 $x$로 넘어가도록 설계했다. 하지만 이렇게 하면 $x=M-x$이고 등장 횟수가 $1$인 경우에도 카운팅을 $1$ 올리므로 틀린 결과를 도출하게 된다

````python
from collections import Counter


def solution():
    N = int(input())
    M = int(input())
    array = list(map(int, input().split()))
    d = Counter(array)
    seen = set()
    for x, c in d.items():
        if M - x not in d:
            continue
        if x == M - x and c == 1:
            continue
        min_, max_ = min(x, M - x), max(x, M - x)
        seen.add((min_, max_))
    answer = len(seen)
    print(answer)


solution()

# input
# 2
# 0
# 0 0
````

`-` 서로 다른 $N$개의 수인 걸 문제 다 풀고 알았다..

## 피보나치수열의 합

- 문제 출처: [정올 3006번](https://jungol.co.kr/problem/3006)

### 정답 풀이 (동적 계획법)

`-` 동적 계획법으로 $1\sim N$번째 피보나치 수를 $O(N)$에 계산할 수 있다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(N)$이고 $N$이 최대 $10^6$이므로 제한 시간 안에 동작할 수 있다

`-` $5$분 걸렸고 $1$번에 맞혔다

````python
def solution():
    N, M = map(int, input().split())
    curr = 1 % M
    prev = 0
    answer = curr
    for i in range(2, N + 1):
        curr, prev = (curr + prev) % M, curr % M
        answer += curr
        answer %= M
    print(answer)


solution()

# input
# 5 11
````

## 두루마리 휴지

- 문제 출처: [정올 8668번](https://jungol.co.kr/problem/8668)

### 정답 풀이(그리디, 우선순위 큐)

`-` $1$번 휴지부터 $N$번 휴지까지 순회하며 사용할지 말지 결정할 것이다. 흡수력 합을 $T$ 이상으로 만들기 위해 흡수력이 음수가 아니면 사용하고 음수라면 사용하지 말자

`-` 여태까지 사용한 휴지들의 흡수력 합이 $T$ 이상인 경우를 고려해보자

`-` 만약 사용한 휴지가 사용 안 한 휴지보다 많다면 사용한 휴지 중 흡수력이 가장 낮은 것을 사용하지 않도록 고려할 수 있다. 이를 흡수력 합이 $T$ 이상이고 개수의 대소 관계가 그대로인 동안 반복해서 적용하자

`-` 이제 반대로 사용 안 한 휴지가 사용한 휴지보다 많다면 사용 안 한 휴지 중 흡수력이 가장 높은 것을 사용하도록 고려하자. 마찬가지로 반복 적용하면 된다

`-` 최솟값과 최댓값은 우선순위 큐를 활용하면 $O(\log N)$에 찾을 수 있다. 이러한 과정을 거치며 문제의 정답을 추적하면 된다

`-` 사용한 휴지들의 흡수력의 합이 $T$ 이상인 상황에서 개수의 균형이 맞지 않으면 최악의 경우 $O(N)$번의 삽입, 삭제가 발생한다

`-` 하지만 균형이 맞는 이후엔 $O(1)$번의 삽입, 삭제만 발생하며 다시 $O(N)$번의 삽입, 삭제가 발생하려면 $O(N)$번의 원소 순회가 필요하다 

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(N\log N)$이다

`-` $33$분 걸렸고 $3$번 만에 맞혔다

````python
import heapq


def solution():
    N, T = map(int, input().split())
    array = list(map(int, input().split()))
    inf = float("inf")
    used_heap = []
    unused_heap = []
    power = 0
    answer = inf
    for s_i in array:
        if s_i >= 0:
            heapq.heappush(used_heap, s_i)
            power += s_i
        else:
            heapq.heappush(unused_heap, -s_i)
        if power < T:
            continue
        while len(used_heap) > len(unused_heap) and power - used_heap[0] >= T:
            s_used_min = heapq.heappop(used_heap)
            power -= s_used_min
            heapq.heappush(unused_heap, -s_used_min)
        while len(unused_heap) > len(used_heap) and power - unused_heap[0] >= T:
            s_unused_min = -1 * heapq.heappop(unused_heap)
            power += s_unused_min
            heapq.heappush(used_heap, s_unused_min)
        answer = min(max(len(used_heap), len(unused_heap)), answer)
    if answer == inf:
        answer = -1
    print(answer)


solution()

# input
# 2 11
# 19 -8
````

## 벌써 12시 (Clock Tree)

- 문제 출처: [정올 5299번](https://jungol.co.kr/problem/5299)

### 정답 풀이 (트리, 그리디)

`-` 되게 난해한 것 같다. 어떻게 해야 할지 고민을 많이 했다

`-` 방과 복도를 노드와 간선으로 생각하면 전체 구조는 트리가 된다

`-` 임의의 간선을 선택하면 해당 간선과 연결된 두 방의 시계를 한 위치 앞당기는 것으로 생각하자

`-` 즉, 한 방에서 시작해 복도를 통해 왔다 갔다 하는 것이다 

`-` 이제 각 방을 루트 노드로 생각한 뒤 루트 노드에서 시작해 모든 방의 시계가 $12$를 가리킬 수 있는지 판단할 것이다

`-` 잘 생각해보면 리프 노드는 간선이 하나뿐이므로 리프 노드 방의 시계가 $x$를 가리키면 해당 방을 $12-x$번 방문해야만 한다

`-` 그러면 리프 노드와 연결된 방의 시간도 같이 앞당겨지며 이제 리프 노드 방의 시계는 고정이므로 이를 제외하고 생각하면 리프 노드와 연결된 노드가 새로운 리프 노드가 된다

`-` 이를 루트 노드까지 재귀적으로 수행한 뒤 루트 노드 방의 시계가 $12$를 가리킨다면 적하는 원하는 바를 이룰 수 있다

`-` 이때 루트 노드에서 시작해 루트 노드에서 방문이 끝날 수도 있지만 그렇지 않을 수도 있다

`-` 이 경우 루트 노드 방의 시계가 $1$시를 가리키고 있게 된다

`-` 결과적으로 모든 방문을 끝마친 뒤 루트 노드 방의 시계가 $12$ 또는 $1$을 가리킨다면 적하는 모든 시계가 $12$를 가리키도록 만들 수 있다

`-` 참고로 시계가 $12$를 가리키면 $0$번의 조작이 필요하고 $1$을 가리키면 $11$번의 조작이 필요하다 (시계가 $12$를 가리키도록 만들기 위해 필요한 방문 횟수)

`-` 각 방을 루트 노드로 생각하고 DFS를 반복해야 하므로 전체 알고리즘의 시간 복잡도는 $O\left(N^2\right)$이다

`-` $52$분 걸렸고 $2$번째 시도에 맞혔다

````python
def dfs(tree, node, parent, clocks):
    x = clocks[node]
    for child in tree[node]:
        if child == parent:
            continue
        need = dfs(tree, child, node, clocks)
        x += need
    return -x % 12


def solution():
    N = int(input())
    clocks = [0] + list(map(int, input().split()))
    tree = [[] for _ in range(N + 1)]
    for _ in range(N - 1):
        u, v = map(int, input().split())
        tree[u].append(v)
        tree[v].append(u)
    none = -1
    answer = 0
    for root in range(1, N + 1):
        need = dfs(tree, root, none, clocks)
        if need == 0 or need == 11:
            answer += 1
    print(answer)
    
    
solution()

# input
# 2
# 10 10
# 1 2
````

## 최대신장트리

- 문제 출처: [정올 1350번](https://jungol.co.kr/problem/1350)

### 정답 풀이 (최소 스패닝 트리)

`-` 최소 스패닝 트리는 크루스칼 알고리즘 사용해서 $O(M\log M)$에 얻을 수 있다

`-` 최대 스패닝 트리를 얻기 위해선 간선의 가중치가 큰 것부터 고려하면 된다. 즉, 정렬 기준만 오름차순에서 내림차순으로 변경하면 된다

`-` $11$분 걸렸고 $1$번에 맞혔다

`-` 최소 스패닝 트리 문제 티어에 비해 구현량이 꽤 있다 (분리 집합 때문에 그럼)

````python
def kruskal(weights, parents, ranks):
    edges = list(weights.items())
    edges.sort(key=lambda e: -e[1])
    mst_weight = 0
    for (u, v), w in edges:
        if find(u, parents) == find(v, parents):
            continue
        union(u, v, parents, ranks)
        mst_weight += w
    return mst_weight


def find(u, parents):
    if u != parents[u]:
        parents[u] = find(parents[u], parents)
    return parents[u]


def union(u, v, parents, ranks):
    u_root = find(u, parents)
    v_root = find(v, parents)
    if u_root == v_root:
        return
    if ranks[u_root] < ranks[v_root]:
        parents[u_root] = v_root
    elif ranks[v_root] < ranks[u_root]:
        parents[v_root] = u_root
    else:
        parents[v_root] = u_root
        ranks[u_root] += 1


def solution():
    N, M = map(int, input().split())
    weights = {}
    for _ in range(M):
        a, b, c = map(int, input().split())
        if b < a:
            a, b = b, a
        weights[a, b] = max(c, weights.get((a, b), 0))
    parents = [i for i in range(N + 1)]
    ranks = [0] * (N + 1)
    answer = kruskal(weights, parents, ranks)
    print(answer)


solution()

# input
# 2 1
# 1 2 3
````

`1.` 어차피 정렬한 후 앞에서부터 union을 수행하므로 최대 가중치를 취하기 위해 간선 관리를 딕셔너리로 할 필요가 없다

`2.` 두 노드가 같은 그룹에 속했는지 확인하기 위해 union 함수에서 불리언 값을 반환하도록 만들자 (find 함수 중복 호출 방지)

## 꿀벌의 이동

- 문제 출처: [정올 1245번](https://jungol.co.kr/problem/1245)

### 정답 풀이 (동적 계획법)

`-` 동적 계획법으로 풀면 되는데 그래프를 어떻게 설계해야 할지 고민을 조금 했다

`-` 시작 좌표는 $(0,0)$으로 설정하자

`-` 현재 좌표가 $(x,y)$일 때 다음 좌표로 가능한 경우를 $(x+1,y), (x-1,y), (x+0.5,1), (x-0.5, 1) ,(x+0.5,-1), (x-0.5, -1)$로 설정하자

`-` 시작 지점에서 아무리 멀리 가도 $x,y$ 좌표값의 절댓값은 $n$을 넘지 못하고 좌표값은 정수 $i$에 대해 $i$ 또는 $i + 0.5$의 형태이므로 가능한 상태 공간의 개수는 $O\left(n^2\right)$이다

`-` 총 $n$번 이동하므로 전체 알고리즘의 시간 복잡도는 $O\left(n^3\right)$이며 $n$은 최대 $20$이므로 제한 시간 안에 동작할 수 있다 

`-` $15$분 걸렸고 $1$번에 맞혔다

````python
from collections import defaultdict


def solution():
    n = int(input())
    dp = {(0, 0): 1}
    dxy = [(1, 0), (-1, 0), (0.5, 1), (-0.5, 1), (0.5, -1), (-0.5, -1)]
    for _ in range(n):
        temp = defaultdict(int)
        for x, y in dp:
            c = dp[x, y]
            for dx, dy in dxy:
                nx, ny = x + dx, y + dy
                temp[nx, ny] += c
        dp = temp
    answer = dp[0, 0]
    print(answer)


solution()

# input
# 9
````

## 여왕벌

- 문제 출처: [정올 2914번](https://jungol.co.kr/problem/2914)

### 정답 풀이 (차분 배열 트릭, 누적 합, 그리디)

`-` 단순 시뮬레이션은 $O\left(NM^2\right)$으로 시간 초과이다

`-` 애벌레가 자라는 수치를 기록한 $2M-1$ 크기의 배열은 감소하지 않는 수열이다

`-` 따라서 $N$일 동안 해당 배열만 갱신하여 한 번에 이를 반영해줘도 된다

`-` 즉, $N$일 동안 누적시킨 $1$열과 $1$행의 자라는 정도를 바탕으로 모든 애벌레가 자라는 정도를 결정하는 것이다

`-` 이것의 시간 복잡도는 $O\left(NM + M^2\right)$이다

`-` 하지만 안타깝게도 $M$은 최대 $700$, $N$은 최대 $1000000$이라 이 역시 시간 초과이다

`-` 하루에 애벌레가 자라는 정도는 $0,1,2$ 중에 하나이다. 따라서 차분 배열을 이용하면 $O(NM)$ 대신 $O(N)$에 이를 수행할 수 있다

`-` 차분 배열을 기반으로 누적 합 배열을 만들면 이것이 $N$일 동안 누적시킨 $1$열과 $1$행의 자라는 정도가 된다

`-` 처음에도 언급했지만 이것은 감소하지 않는 수열이다. 또한, 나머지 애벌레들은 자신의 왼쪽, 왼쪽 위, 위쪽의 애벌레들이 다 자란 다음, 그날 가장 많이 자란 애벌레가 자란 만큼 자신도 자란다

`-` 그러므로 나머지 애벌레들은 자신의 위쪽 애벌레가 자란 만큼 자신도 자란다 

`-` 전체 알고리즘의 시간 복잡도는 $O\left(N + M^2\right)$이다

`-` $1$시간 $25$분 걸렸고 $3$번 만에 맞혔다

`-` 차분 배열을 구상하는데 너무 오래걸렸다 (인덱스 설정도 이상하게 해서 계속 틀렸음)

````python
def make_difference_array(growth_values_per_day):
    n = sum(growth_values_per_day[0])
    difference_array = [0] * n
    for n_zeros, _, n_twos in growth_values_per_day:
        if n_zeros == n:
            continue
        difference_array[n_zeros] += 1
        if n_twos > 0:
            difference_array[n - n_twos] += 1
    return difference_array


def compute_prefix_sums(array):
    n = len(array)
    prefix_sums = [1] * n
    value = 0
    for i in range(n):
        value += array[i]
        prefix_sums[i] += value
    return prefix_sums


def solution():
    M, N = map(int, input().split())
    growth_values_per_day = [list(map(int, input().split())) for _ in range(N)]
    difference_array = make_difference_array(growth_values_per_day)
    growth_values = compute_prefix_sums(difference_array)
    first_col = growth_values[:M][::-1]
    first_row = growth_values[M : 2 * M]
    for i in range(M):
        print(first_col[i], *first_row)


solution()

# input
# 1 1
# 0 0 1
````

## 풍선 터트리기(BALONI)

- 문제 출처: [정올 2978번](https://jungol.co.kr/problem/2978)

### 시간 초과 풀이 (그리디, 시뮬레이션, 세그먼트 트리)

`-` 만약 화살의 높이가 줄어들지 않는다면 정답은 유니크한 풍선의 높이 개수가 된다. 하지만 문제는 그렇지 않다

`-` 현재 가장 높이 떠있는 풍선의 높이를 $h_i$라 하자. 이 풍선을 터트리기 위해선 높이 $h_i$에서 화살을 발사해야만 한다

`-` 화살이 $i$번째 풍선을 터트린 뒤 화살의 높이는 $h_i - 1$이 된다. 이 화살은 아직 터지지 않고 높이가 $h_i - 1$인 $i+1 \sim N$번째 풍선 중 가장 인덱스가 앞서는 풍선을 터트린다. 화살의 높이는 $h_i-2$가 되며 화살의 높이가 $0$이 되거나 $N$번째 풍선을 지나칠 때까지 이것이 재귀적으로 반복된다

`-` 단순 순회로 작업하자니 한 번 발사마다 단 하나의 풍선만 터지는 최악의 경우엔 시간 복잡도가 $O\left(N^2\right)$이라 시간 초과이다

`-` 임의의 구간 내 최댓값 중 인덱스가 가장 앞서는 것을 빠르게 찾을 수 있다면 문제를 빠르게 해결할 수 있다. 이는 세그먼트 트리를 사용하면 $O(\log N)$에 수행할 수 있다

`-` 아직 터지지 않은 풍선 중 가장 높으며 이것이 여러 개라면 인덱스가 가장 뒤서는 풍선을 고려하자 (이 풍선의 인덱스가 $i$일 때 높이는 $h_i$이다)

`-` 실제로 터지는 건 인덱스가 가장 앞서는 것이지만 어차피 이들 전부를 터트리기 위해 개수만큼 화살을 발사해야 하므로 순서는 상관없다

`-` 화살은 $i+1\sim N$번째 풍선 중 높이가 $h_i - 1$이며 인덱스가 가장 앞서는 것을 터트리게 된다

`-` 해당 구간에 속한 풍선의 높이는 아무리 높아도 $h_i-1$이다. 따라서 이 구간에서 최댓값의 인덱스 $j$를 찾고 $h_j$가 $h_i-1$인지 확인하면 된다

`-` 동일하다면 화살을 해당 풍선을 터트리고 높이가 $h_i-2$가 되며 작업을 재귀적으로 반복하면 된다. 그렇지 않다면 새로운 화살을 발사해야 한다

`-` 전체 알고리즘의 시간 복잡도는 최악의 경우에도 $O(N\log N)$이 되므로 제한 시간 안에 동작할 수 있다

`-` PyPy는 메모리 초과이고 Python은 시간 초과이다. 최적화를 해봤지만 PyPy도 시간 초과가 발생했다

`-` $N$이 최대 $10^6$이고 파이썬이 느려서 빠듯한 것 같다. 다른 방법이 있는지 생각해 보자

```python
import math


def simulate(heights):
    n = len(heights)
    h = math.ceil(math.log2(n + 1))
    tree_size = 2**(h + 1)
    tree = [n - 1] * tree_size
    root = 1
    start, end = 0, n - 1
    init(heights, tree, root, start, end)
    left, right = 0, n - 1
    count = 0
    while heights[tree[root]] > 0:
        count += 1
        i = query(left, right, heights, tree, root, start, end)
        h_i = heights[i]
        update(heights, i, 0, tree, root, start, end)
        while True:
            j = query(i + 1, right, heights, tree, root, start, end)
            if j is None:
                break
            h_j = heights[j]
            if h_j + 1 != h_i:
                break
            update(heights, j, 0, tree, root, start, end)
            i = j
            h_i = h_j
    return count


def init(array, tree, node, start, end):
    if start == end:
        tree[node] = start
        return
    mid = (start + end) // 2
    init(array, tree, 2 * node, start, mid)
    init(array, tree, 2 * node + 1, mid + 1, end)
    h_l = array[tree[2 * node]]
    h_r = array[tree[2 * node + 1]]
    if h_l > h_r:
        tree[node] = tree[2 * node]
    else:
        tree[node] = tree[2 * node + 1]


def update(array, index, value, tree, node, start, end):
    if index < start or index > end:
        return
    if start == end:
        array[index] = value
        return
    mid = (start + end) // 2
    update(array, index, value, tree, 2 * node, start, mid)
    update(array, index, value, tree, 2 * node + 1, mid + 1, end)
    h_l = array[tree[2 * node]]
    h_r = array[tree[2 * node + 1]]
    if h_l > h_r:
        tree[node] = tree[2 * node]
    else:
        tree[node] = tree[2 * node + 1]


def query(left, right, array, tree, node, start, end):
    if left > right:
        return None
    if left > end or right < start:
        return INF
    if left <= start and end <= right:
        return tree[node]
    mid = (start + end) // 2
    left_argmax = query(left, right, array, tree, 2 * node, start, mid)
    right_argmax = query(left, right, array, tree, 2 * node + 1, mid + 1, end)
    if left_argmax == INF:
        return right_argmax
    if right_argmax == INF:
        return left_argmax
    h_l = array[left_argmax]
    h_r = array[right_argmax]
    if h_l > h_r:
        return left_argmax
    return right_argmax


def solution():
    global INF
    INF = float("inf")
    N = int(input())
    heights = list(map(int, input().split()))
    heights.append(0)
    answer = simulate(heights)
    print(answer)


solution()
```

### 정답 풀이 (그리디, 시뮬레이션, 스택)

`-` 높이별로 풍선들의 인덱스를 저장해두자. 높이가 같은 것끼리 묶은 걸 원소로 가지는 배열을 생성하고 높이를 기준으로 오름차순 정렬하자. 원소도 인덱스를 기준으로 오름차순 정렬하자 

`-` 가장 높은 풍선은 배열의 마지막 원소에 존재한다

`-` 마지막 원소에 저장된 풍선들의 인덱스 중 마지막 원소를 제거하자. 이는 $O(1)$이다

`-` 만약 마지막 원소에 저장된 인덱스가 비어있다면 마지막 원소를 pop하고 새롭게 바뀐 마지막 원소에 대해 재귀적으로 수행하면 된다

`-` 이때 마지막 원소의 인덱스가 $m$이고 제거한 풍선의 인덱스가 $i$라 하자

`-` 만약 $m-1$번째 원소에 저장된 풍선들의 높이가 $h_i-1$이 아니라면 화살을 새롭게 발사해야 한다

`-` 그렇지 않다면 마지막 원소를 확인하자. 인덱스가 $i$보다 크다면 제거할 수 있고 위의 과정을 재귀적으로 반복하면 된다

`-` 전체 알고리즘의 시간 복잡도는 $O(N)$이다

`-` 총 $3$시간 $24$분 걸렸고 $8$번 만에 맞혔다......

````python
from collections import defaultdict


def simulate(array):
    count = 0
    while len(array) >= 1:
        n = len(array)
        if not array[n - 1][1]:
            array.pop()
            continue
        h_i = array[n - 1][0]
        i = array[n - 1][1].pop()
        count += 1
        k = n - 2
        while k >= 0:
            if not array[k][1]:
                break
            h_j = array[k][0]
            j = array[k][1][-1]
            if h_j != h_i - 1 or i > j:
                break
            array[k][1].pop()
            h_i = h_j
            i = j
            k -= 1
    return count


def solution():
    N = int(input())
    heights = list(map(int, input().split()))
    height2indices = defaultdict(list)
    for i, h in enumerate(heights):
        height2indices[h].append(i)
    array = list(height2indices.items())
    array.sort()
    for h, indices in array:
        indices.sort()
    answer = simulate(array)
    print(answer)


solution()

# input
# 4
# 4 3 1 2
````

## 주식거래자

- 문제 출처: [정올 1748번](https://jungol.co.kr/problem/1748)

### 정답 풀이 (다익스트라, 브루트 포스)

`-` 주식거래자를 노드로 생각하고 정보의 전달을 간선으로 표현하자. 이때 간선의 가중치는 정보를 전달하는데 걸리는 시간이다

`-` 임의의 시작 노드에 대해 다익스트라 알고리즘을 수행하자. 이러면 모든 사람에게 정보를 전달하는데 걸리는 최소 시간을 알 수 있다. 만약 이것이 $\infty$이면 지금의 시작 노드에선 모든 사람에게 정보를 전달할 수 없는 것이다

`-` 각 노드마다 다익스트라 알고리즘을 수행하면 정답을 알 수 있다

`-` $29$분 걸렸고 $2$번째 시도에 맞혔다 (변수명 실수)

````python
import heapq
from collections import defaultdict


def minimize_time(graph, weights):
    m = len(weights) - 1
    start = 1
    min_time = INF
    for u in range(1, m + 1):
        distances = dijkstra(graph, weights, u)
        max_ = max(distances)
        if min_time <= max_:
            continue
        start = u
        min_time = max_
    return start, min_time


def dijkstra(graph, weights, start):
    m = len(weights) - 1
    queue = [(0, start)]
    distances = [INF] * (m + 1)
    distances[0] = 0
    distances[start] = 0
    while queue:
        d_u, u = heapq.heappop(queue)
        if distances[u] < d_u:
            continue
        for v in graph[u]:
            w = weights[u][v]
            d_v = d_u + w
            if d_v >= distances[v]:
                continue
            distances[v] = d_v
            heapq.heappush(queue, (d_v, v))
    return distances


def solve_testcase(m):
    graph = defaultdict(list)
    weights = [[0] * (m + 1) for _ in range(m + 1)]
    for u in range(1, m + 1):
        data = list(map(int, input().split()))
        n = data[0]
        for i in range(1, n + 1):
            v, t = data[2 * i - 1], data[2 * i]
            graph[u].append(v)
            weights[u][v] = t
    start, min_time = minimize_time(graph, weights)
    if min_time == INF:
        print("disjoint")
        return
    print(start, min_time)


def solution():
    global INF
    INF = float("inf")
    while True:
        m = int(input())
        if m == 0:
            break
        solve_testcase(m)


solution()

# input
# 0
````

## 보물 마차

- 문제 출처: [정올 12097번](https://jungol.co.kr/problem/12097)

### 정답 풀이 (그리디, 우선순위 큐)

`-` 각 마차에는 최대 $1$개의 보물만 실을 수 있다. 따라서 자기가 실을 수 있는 보물 중 가치가 최대인 것을 실어야 이득이다

`-` 실을 수 있는 무게별로 마차를 정렬하자. 그리고 보물도 무게를 기준으로 정렬하자

`-` 각 마차를 순회하며 현재 자기가 실을 수 있는 보물을 고려하자

`-` 이 중 가치가 최대인 것을 골라야 하는데 단순 순회는 $O(N)$이다. 대신 우선순위 큐를 사용해서 $O(\log N)$에 수행하자

`-` 무게별로 마차를 정렬했기에 이전의 마차들이 실을 수 있는 보물은 현재 마차에도 실을 수 있다

`-` 따라서 현재 마차에 실을 수 있는 보물을 모두 고려한 뒤 최대 가치의 보물을 고르는 전략은 항상 유효하다

`-` 전체 알고리즘의 시간 복잡도는 $O(N\log N + M\log M)$이다

`-` $11$분 걸렸고 $1$번에 맞혔다

````python
import heapq


def solution():
    N, M = map(int, input().split())
    treasures = [list(map(int, input().split())) for _ in range(N)]
    treasures.sort(reverse=True)
    carriages = list(map(int, input().split()))
    carriages.sort()
    total_value = 0
    queue = []
    for s in carriages:
        while treasures and treasures[-1][0] <= s:
            w, v = treasures.pop()
            heapq.heappush(queue, -v)
        if not queue:
            continue
        v = heapq.heappop(queue)
        v *= -1
        total_value += v
    print(total_value)


solution()

# input
# 1 1
# 1 1
# 1
````

## 이동 방해

- 문제 출처: [정올 3876번](https://jungol.co.kr/problem/3876)

### 정답 풀이 (다익스트라, 역추적, 그리디)

`-` 다익스트라 알고리즘을 $M$번 반복하면 정답을 찾을 수 있다. 문제는 시간 복잡도가 $O\left(M^2 \log M\right)$이고 $M$이 최대 $25000$이라 시간 초과라는 것이다

`-` 잘 생각해보면 모든 $M$개의 간선에 대해 가중치를 기존의 $2$배로 바꿔가며 다익스트라 알고리즘을 수행할 필요가 없다

`-` 다익스트라 알고리즘을 최초 한 번 수행하자. 연결 그래프이므로 항상 경로가 존재하며 이 경로에 포함된 간선의 개수는 최대 $N-1$이다

`-` 최단 경로에 포함된 간선 말고 다른 간선의 가중치를 $2$배로 늘리는 건 아무 의미가 없다. 어차피 해당 간선을 사용하지 않기 때문이다

`-` 따라서 $M$개 대신 최대 $N-1$개만 조작해도 되므로 전체 알고리즘의 시간 복잡도는  $O\left(NM \log M\right)$이 된다

`-` $26$분 걸렸고 $1$번에 맞혔다

````python
import heapq


def dijkstra(graph, weights, start):
    n = len(graph)
    queue = [(0, start)]
    distances = [INF] * n
    distances[start] = 0
    predecessors = [0] * n
    while queue:
        d_u, u = heapq.heappop(queue)
        if distances[u] < d_u:
            continue
        for v in graph[u]:
            w = weights[u][v]
            d_v = d_u + w
            if distances[v] <= d_v:
                continue
            distances[v] = d_v
            predecessors[v] = u
            heapq.heappush(queue, (d_v, v))
    return distances, predecessors


def track(start, end, predecessors):
    path = []
    node = end
    while node != start:
        path.append(node)
        node = predecessors[node]
    path.append(start)
    return path


def maximize_shortest_time(graph, weights, start, end):
    original_distances, predecessors = dijkstra(graph, weights, start)
    path = track(start, end, predecessors)
    original_shortest_time = original_distances[end]
    gap = 0
    for i in range(len(path) - 1):
        u, v = path[i], path[i + 1]
        w = weights[u][v]
        weights[u][v] = weights[v][u] = 2 * w
        distances, _ = dijkstra(graph, weights, start)
        weights[u][v] = weights[v][u] = w
        gap = max(distances[end] - original_shortest_time, gap)
    return gap


def solution():
    global INF
    INF = float("inf")
    N, M = map(int, input().split())
    graph = [[] for _ in range(N + 1)]
    weights = [[0] * (N + 1) for _ in range(N + 1)]
    for _ in range(M):
        A, B, L = map(int, input().split())
        graph[A].append(B)
        graph[B].append(A)
        weights[A][B] = weights[B][A] = L
    answer = maximize_shortest_time(graph, weights, 1, N)
    print(answer)


solution()

# input
# 2 1
# 1 2 3
````

## List of Unique Integers

- 문제 출처: [정올 3017번](https://jungol.co.kr/problem/3017)

### 정답 풀이 (큐, 해시를 사용한 집합과 맵)

`-` $n$번째 원소까지 고려하여 가능한 연속 부분 수열의 경우의 수를 계산했다고 해보자

`-` 이때 $n+1$번째 원소인 $a_{n+1}$을 추가로 고려했을 때 경우의 수가 어떻게 되는지 보자

`-` 일단 $a_{n+1}$을 연속 부분 수열에 포함시키지 않는다면 의미가 없으므로 $a_{n+1}$은 연속 부분 수열에 포함시켜야 한다

`-` 그럼 $a_{n+1}$로 인해 새롭게 고려되는 연속 부분 수열은 $\{a_{n+1}\}, \{a_n, a_{n+1}\} ,  \{a_{n-1}, a_n, a_{n+1}\} ,\dots, \{a_i, \dots, a_{n-1}, a_n, a_{n+1}\}$와 같다 (단, $a_{i-1} = a_{n+1}$이며 연속 부분 수열의 모든 원소는 고유하다)

`-` 최장 연속 부분 수열의 길이는 $n-i+2$이다. 따라서 새롭게 추가된 연속 부분 수열의 개수는 $n-i+2$이다

`-` $n$번째 원소를 마지막 원소로 가지는 최장 연속 부분 수열을 고려하자. 그리고 해당 연속 부분 수열에 존재하는 고유 원소 집합을 가지고 있자

`-` 이제 $n+1$번째 원소를 최장 연속 부분 수열에 추가하려고 시도할 것이다

`-` 만약, $a_{n+1}$이 고유 원소 집합에 존재하지 않는다면 $a_{n+1}$을 문제 없이 추가할 수 있고 최장 연속 부분 수열의 길이만큼 경우의 수가 누적된다

`-` 이제 그렇지 않다고 해보자. 그럼 최장 연속 부분 수열의 왼쪽부터 $a_{n+1}$과 같은 값을 제거할 때까지 popleft를 수행하자

`-` 그리고 $a_{n+1}$을 추가한 뒤 길이만큼 경우의 수를 누적하면 된다

`-` queue로 관리하면 popleft를 $O(1)$에 수행할 수 있으며 모든 원소는 연속 부분 수열에 $1$번만 포함되고 최대 $1$번만 제거된다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(N)$이다

`-` $23$분 걸렸고 $2$번 만에 맞혔다

`-` 원소를 popleft하면 고유 원소 집합에서도 제거해줘야 하는데 빼먹어서 $1$번 틀렸다

````python
from collections import deque


def solution():
    N = int(input())
    array = list(map(int, input().split()))
    count = 0
    subsequence = deque([])
    unique = set()
    for a in array:
        if a not in unique:
            unique.add(a)
        else:
            while subsequence[0] != a:
                x = subsequence.popleft()
                unique.remove(x)
            subsequence.popleft()
        subsequence.append(a)
        count += len(subsequence)
    print(count)


solution()
````

## 트리의 지름

- 문제 출처: [정올 4947번](https://jungol.co.kr/problem/4947)

### 정답 풀이 (트리, 그래프 탐색)

`-` 웰노운 문제이다. 임의의 정점에서 DFS를 수행하여 가장 먼 노드 $u$를 찾자. 트리이므로 BFS가 아닌 DFS로도 최장 거리를 찾을 수 있다

`-` 이때 노드 $u$에서 가장 멀리 떨어진 노드 $v$와의 거리가 트리의 지름이 된다

`-` 노드 $v$도 DFS를 수행해서 찾으면 되며 DFS를 $2$번 수행하므로 전체 알고리즘의 시간 복잡도는 $O(n)$이다

`-` $12$분 걸렸고 $1$번에 맞혔다

````python
def find_diameter(tree):
    n = len(tree)
    none, root = 0, 1
    distances = [-1] * n
    dfs(tree, root, none, distances)
    farthest_node_from_root = find_argmax(distances)
    distances = [-1] * n
    dfs(tree, farthest_node_from_root, none, distances)
    diameter = max(distances)
    return diameter


def dfs(tree, node, parent, distances):
    distances[node] = distances[parent] + 1
    for child in tree[node]:
        if child == parent:
            continue
        dfs(tree, child, node, distances)


def find_argmax(array):
    max_ = max(array)
    for i, a in enumerate(array):
        if a == max_:
            return i


def solution():
    n = int(input())
    tree = [[] for _ in range(n + 1)]
    for _ in range(n - 1):
        a, b = map(int, input().split())
        tree[a].append(b)
        tree[b].append(a)
    answer = find_diameter(tree)
    print(answer)


solution()

# input
# 2
# 1 2
````

## 문자들(LETTERS)

- 문제 출처: [정올 3529번](https://jungol.co.kr/problem/3529)

### 정답 풀이 (백트래킹)

`-` 이동 규칙 중 $1$번은 자명하며 $2$번은 $3$번에 종속된다. 따라서 $3$번 규칙만 유의하면 된다

`-` 각 셀에는 대문자 알파벳만이 들어있다. 알파벳은 총 $26$개이니 최대로 얻을 수 있는 상품 개수도 $26$이다

`-` 여태까지 획득한 상품의 상태 공간을 집합으로 관리하며 셀을 이동하자. 이때 백트래킹을 사용해 $3$번 규칙에 따라 가지치기를 수행하면 탐색 시간을 대폭 줄일 수 있다

`-` 솔직히 최악의 경우에서 동작하는 알고리즘의 시간 복잡도는 모르겠다

`-` 참고로 왼쪽 상단은 시작 셀로 무조건 밟아야 함에 유의하자

`-` $21$분 걸렸고 $1$번에 맞혔다

````python
def backtracking(graph, r, c, n_rows, n_cols, drc, seen):
    global ANSWER
    n = len(seen)
    if n > ANSWER:
        ANSWER = n
    for dr, dc in drc:
        nr, nc = r + dr, c + dc
        is_in_range = 0 <= nr < n_rows and 0 <= nc < n_cols
        if not is_in_range:
            continue
        a = graph[nr][nc]
        if a in seen:
            continue
        seen.add(a)
        backtracking(graph, nr, nc, n_rows, n_cols, drc, seen)
        seen.remove(a)


def solution():
    global ANSWER
    R, C = map(int, input().split())
    board = [input() for _ in range(R)]
    drc = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    ANSWER = 0
    backtracking(board, 0, 0, R, C, drc, set(board[0][0]))
    print(ANSWER)


solution()
````

## 하노이6(초보들의 하노이)

- 문제 출처: [정올 3272번](https://jungol.co.kr/problem/3272)

### 정답 풀이 (동적 계획법)

`-` 각 원판이 어느 기둥에 위치하고 있는지와 현재 몇 번째 이동인지가 동일하면 현재 상태를 만드는 경우의 수 역시 동일하다

`-` 각 원판이 위치할 수 있는 기둥은 $3$개이므로 가능한 원판 배치의 수는 $O\left(3^N\right)$이다. $2$번 규칙에 의해 실제 경우의 수는 시간 복잡도에 비해 훨씬 작다

`-` 또, 각 기둥별로 가장 위의 있는 원판은 $O(N)$에 알 수 있으며 고유한 상태 공간을 만들기 위해 각 기둥별로 쌓여진 원판 번호를 정렬하여 가지고 있을 것이다. 이는 $O(N \log N)$이다

`-` 따라서 동적 계획법을 이용하면 $O\left(NM 3 ^ N \log N \right)$에 문제를 해결할 수 있다. 점화식을 전개할 때 $3$번 규칙에 유의하자

`-` $N$은 최대 $5$이고 $M$은 최대 $1000$이므로 제한 시간 안에 충분히 실행될 수 있다

`-` $43$분 걸렸고 $1$번에 맞혔다

````python
from collections import defaultdict


def hanoi(n, m, mod):
    dp = [defaultdict(int) for _ in range(m + 2)]
    dp[0][tuple(range(1, n + 1)), (), ()] = 1
    none = n + 1
    num_ways = 0
    for i in range(m + 1):
        for (first, second, third), count in dp[i].items():
            dp[i][first, second, third] %= mod
            if len(third) == n:
                num_ways += count
                num_ways %= mod
                continue
            min_first = min(first, default=none)
            min_second = min(second, default=none)
            min_third = min(third, default=none)
            state = {"first": first, "second": second, "third": third}
            if min_first < min_second:
                update(count, first, second, min_first, min_second, "first", "second", state.copy(), dp[i + 1])
            if min_first < min_third:
                update(count, first, third, min_first, min_third, "first", "third", state.copy(), dp[i + 1])
            if min_second < min_first:
                update(count, second, first, min_second, min_first, "second", "first", state.copy(), dp[i + 1])
            if min_second < min_third:
                update(count, second, third, min_second, min_third, "second", "third", state.copy(), dp[i + 1])
            if min_third < min_first:
                update(count, third, first, min_third, min_first, "third", "first", state.copy(), dp[i + 1])
            if min_third < min_second:
                update(count, third, second, min_third, min_second, "third", "second", state.copy(), dp[i + 1])
    return num_ways


def update(count, from_, to, min_from, min_to, from_type, to_type, state, dp):
    if min_from >= min_to:
        return
    _from, _to = list(from_), list(to)
    _from.remove(min_from)
    _to.append(min_from)
    _to.sort()
    state[from_type], state[to_type] = tuple(_from), tuple(_to)
    dp[tuple(state.values())] += count


def solution():
    N, M = map(int, input().split())
    mod = 10**9 + 7
    answer = hanoi(N, M, mod)
    print(answer)


solution()

# input
# 3 8
````

`-` 코드가 너무 더러워서 리팩토링 좀 했다

`-` $22$분 걸린 듯 (풀이 시간에 포함 안 시킴)

- 재미나이 피셜

`-` 원판을 옮길 때 기둥의 맨 위에 얹혀진다. 따라서 정렬은 $O(N\log N)$이 아니라 $O(N)$이다 (내부적으로 Timsort 사용, 맨 앞에 원소 하나 추가)

`-` 가능한 원판 배치의 수가 $O\left(\min\left(3^N, 3^M\right)\right)$이며 각 배치를 관리하는데 $O(N)$이 필요하므로 상태 공간의 크기는 엄밀하게 $O\left(N \cdot \min \left(3^N, 3^M\right)\right)$이다

## 팰린드롬 유형

- 문제 출처: [정올 11236번](https://jungol.co.kr/problem/11236)

### 정답 풀이 (애드 혹, 재귀)

`-` 만약 $w$가 $\text{type-$0$}$ 팰린드롬이라면 $w$의 양 끝 문자는 동일하다

`-` 이에 착안해 $w$의 양 끝 문자를 보자. 만약 이 둘이 같다면 양 끝 문자를 없애도 $w$의 $\text{type-$k$}$는 변하지 않는다. 따라서 이 둘을 없앤 뒤 새로운 문자열에 대해 양 끝 문자를 재귀적으로 확인하자. 양 끝의 문자를 없애는 건 문자열의 양 끝을 가리키는 포인터를 반대편으로 $1$칸씩 옮기는 방식으로 구현할 수 있다

`-` 이제 $w$의 양 끝 문자가 다르다고 해보자. 그럼 $w$를 팰린드롬으로 만들기 위해 양 끝 문자 중 적어도 하나는 제거해야 한다. 양 끝 문자 중 하나를 제거한 문자열을 $w'$이라 하자

`-` $w'$이 $\text{type-$0$}$ 팰린드롬이라면 $w$는 $\text{type-$1$}$ 팰린드롬이다. 하지만 $w'$이 $\text{type-$0$}$ 팰린드롬이 아닐 수 있다

`-` 잘 생각해 보면 주어진 문자열이 $w$에서 $w'$으로 바뀌었을 뿐 나머지는 동일하다

`-` 따라서 동일한 논리에 따라 $w'$의 양 끝 문자를 확인해 나가면 된다

`-` 제거할 수 있는 문자 수는 최대 $3$이며 제거 방향은 둘 중 하나이다. 그러므로 가능한 경우의 수는 $2^3$이다

`-` 즉, 아무리 많아도 $8$개의 경우를 확인하면 $w$가 어떤 $\text{type-$k$}$ 팰린드롬인지 알 수 있다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(n)$이다

`-` $25$분 걸렸고 $1$번에 맞혔다

````python
import sys

sys.setrecursionlimit(10**5 + 2)


def check_type_k(s, left, right, n_removals):
    if n_removals > 3:
        return INF
    if left > right:
        return 0
    if s[left] == s[right]:
        return check_type_k(s, left + 1, right - 1, n_removals)
    k1 = check_type_k(s, left + 1, right, n_removals + 1)
    k2 = check_type_k(s, left, right - 1, n_removals + 1)
    return 1 + min(k1, k2)


def solution():
    global INF
    INF = float("inf")
    w = input().rstrip()
    answer = check_type_k(w, 0, len(w) - 1, 0)
    if answer == INF:
        answer = -1
    print(answer)


solution()

# input
# abdeba
````
